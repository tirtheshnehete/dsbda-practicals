```python
def fun1(value):
    if (value=="male"):
        return 1
    elif (value == "female"):
        return 0
    else:
        pass
```

```python
df["Age"].value_counts()
```

```python
def fun2(value):
    if (np.isnan(value)==False):
        return int(value)
```

```python
df["Age"]=df["Age"].apply(fun2)
```

```python
def fun3(value):
    if (value < 20):
        return "teenager"
    elif (value >= 20 and value < 40):
        return "young"
    elif (value >= 40 and value < 60):
        return "middle aged"
    elif (value >= 60):
        return "senior citizen"
    else:
        pass
```

```python
df["Age"]=df["Age"].apply(fun3)
```
